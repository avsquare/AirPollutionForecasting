```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inputData=pd.read_csv("dataset.csv", low_memory=False)
```

```python
min_year= pd.to_datetime(inputData["date"]).apply(lambda x: x.year).min()
```

```python
inputData["time_index"]=((pd.to_datetime(inputData["date"]).apply(lambda x: x.year) - 
 min_year)*12+ 
 pd.to_datetime(inputData["date"]).apply(lambda x: x.month))
```
